```python
# Now we will create a user defined function for finding the lower range and upper range for a variable so that outliers can be treated.
def remove_outliers(col):
    sorted(col)
    Q1, Q3 = col.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
```

```python
lrAI, urAI = remove_outliers(df['ApplicantIncome'])
df['ApplicantIncome'] = np.where(df['ApplicantIncome'] > urAI, urAI, df['ApplicantIncome'])
df['ApplicantIncome'] = np.where(df['ApplicantIncome'] < lrAI, lrAI, df['ApplicantIncome'])

lrCoI, urCoI = remove_outliers(df['CoapplicantIncome'])
df['CoapplicantIncome'] = np.where(df['CoapplicantIncome'] > urCoI, urCoI, df['CoapplicantIncome'])
df['CoapplicantIncome'] = np.where(df['CoapplicantIncome'] < lrCoI, lrCoI, df['CoapplicantIncome'])

lrLA, urLA = remove_outliers(df['LoanAmount'])
df['LoanAmount'] = np.where(df['LoanAmount'] > urLA, urLA, df['LoanAmount'])
df['LoanAmount'] = np.where(df['LoanAmount'] < lrLA, lrLA, df['LoanAmount'])

lrLAT, urLAT = remove_outliers(df['Loan_Amount_Term'])
df['Loan_Amount_Term'] = np.where(df['Loan_Amount_Term'] > urLAT, urLAT, df['Loan_Amount_Term'])
df['Loan_Amount_Term'] = np.where(df['Loan_Amount_Term'] < lrLAT, lrLAT, df['Loan_Amount_Term'])

lrCH, urCH = remove_outliers(df['Credit_History'])
df['Credit_History'] = np.where(df['Credit_History'] > urCH, urCH, df['Credit_History'])
df['Credit_History'] = np.where(df['Credit_History'] < lrCH, lrCH, df['Credit_History'])

df.shape
```

```python
# Replacing null values in numerical columns with median
median_1 = df['ApplicantIncome'].median()
median_2 = df['CoapplicantIncome'].median()
median_3 = df['LoanAmount'].median()
median_4 = df['Loan_Amount_Term'].median()
median_5 = df['Credit_History'].median()

df['ApplicantIncome'].replace(np.nan, median_1 ,inplace = True)
df['CoapplicantIncome'].replace(np.nan, median_2, inplace = True)
df['LoanAmount'].replace(np.nan, median_3, inplace = True)
df['Loan_Amount_Term'].replace(np.nan, median_4, inplace = True)
df['Credit_History'].replace(np.nan, median_5, inplace = True)
```

```python
# Replacing null values in Categorical columns with mode
mode_1 = df['Gender'].mode()[0]
mode_2 = df['Married'].mode()[0]
mode_3 = df['Dependents'].mode()[0]
mode_4 = df['Education'].mode()[0]
mode_5 = df['Self_Employed'].mode()[0]
mode_6 = df['Property_Area'].mode()[0]
mode_7 = df['Loan_Status'].mode()[0]

df['Gender'] = df['Gender'].replace(np.nan, mode_1)
df['Married'] = df['Married'].replace(np.nan, mode_2)
df['Dependents'] = df['Dependents'].replace(np.nan, mode_3)
df['Education'] = df['Education'].replace(np.nan, mode_4)
df['Self_Employed'] = df['Self_Employed'].replace(np.nan, mode_5)
df['Property_Area'] = df['Property_Area'].replace(np.nan, mode_6)
df['Loan_Status'] = df['Loan_Status'].replace(np.nan, mode_7)
```

### Normalizing and Scalling
